```python
import pandas as pd 

f = '파일명'
# f = 'C:/Users/82103/Desktop/Python/project/excel.xlsx'
file = pd.read_excel(f)                     #역대 당첨번호 저장 엑셀 파일 불러오기
df_result = pd.DataFrame(file)              #엑셀파일을 dataframe으로 
df_list = df_result.values.tolist()         #dataframe 리스트로 

# print(df_result.head())
```

```python
# a = input('시작 회차를 입력하세요 : ')        #시작, 끝 회차 지정
# b = input(' 끝 회차를 입력하세요 : ')
a,b = 935 ,0
a = len(df_list)-int(a)
b = len(df_list)-int(b)


  
num = []                                        # DataFrame 만들기 위해서 1부터 45까지 리스트 만들기
for i in range(1,46) : 
    num.append(str(i))

df_matrix=pd.DataFrame(columns=num,index = num) # 45x45 행렬 만들기
df_matrix[:] = df_matrix[:].fillna(0)           # NaN -> 0으로 바꾸기 


for n_th in df_list[a:b+1]:                        #각 회차별로 당첨번호간 상관간계 구하기
    for index_th in range(0,5):                    #각 회차별 번호 하나씩 저장 
        y = str(n_th[index_th])                    # ex) 1 2 3 4 5 6이면, 
        for column_th in range(index_th+1, 6):    # (1,2) +1  , (1,3) , (1,4) , (1,5), (1,6)
            x = str(n_th[column_th])              #(2,3) ,(2,4), (2,5), (2,6),
            df_matrix[y][x] +=1                   # (3,4), (3,5), (3,6), (4,5), (4,6) , (5,6)에 +1 

import seaborn as sns                            
hmap=sns.heatmap(df_matrix)                      #결과 값 히트맵으로 표시
hmap.figure.savefig("로또 분포도.png",
                    format='png',
                    dpi=150)                     #히트맵 저장 


v0 = df_matrix[df_matrix == 26].stack().index.tolist()       #당첨번호 같이 나온 횟수가 26개 이상
v1 = df_matrix[df_matrix == 25].stack().index.tolist()       #당첨번호 같이 나온 횟수가 25개 이상
v2 = df_matrix[df_matrix == 24].stack().index.tolist()
v3 = df_matrix[df_matrix == 23].stack().index.tolist()
v4 = df_matrix[df_matrix == 22].stack().index.tolist()
v5 = df_matrix[df_matrix == 21].stack().index.tolist()

print(v0,'\n', v1 ,'\n', v2,'\n', v3,'\n', v4,'\n')

print(v5)

```

```python
# 935회 - 900회 
a,b = 935 ,900
a  = len(df_list)-int(a)
b = len(df_list)-int(b)


#각 회차별로 당첨번호간 상관간계 구하기

num = []
for i in range(1,46) : 
    num.append(str(i))

df_matrix=pd.DataFrame(columns=num,index = num) 
df_matrix[:] = df_matrix[:].fillna(0)


for n_th in df_list[a:b+1]:                    #각 회차별로 당첨번호간 상관간계 구하기
    for index_th in range(0,5):
        y = str(n_th[index_th])
        for column_th in range(index_th+1, 6):
            x = str(n_th[column_th])
            df_matrix[y][x] +=1 

import seaborn as sns
hmap=sns.heatmap(df_matrix)
hmap.figure.savefig("로또분포도_900_850.png",
                    format='png',
                    dpi=150)


v4 = df_matrix[df_matrix >= 4].stack().index.tolist()
v3 = df_matrix[df_matrix == 3].stack().index.tolist()

print(v4, '\n', v3)
```
